```python
#import all tools
import yfinance as yf #collect data
import pandas as pd #data analysis
import matplotlib.pyplot as plt #data visualisation
import seaborn as sns #data visualisation
```

```python
#define data extraction end and start date from yahoo finance
start='2015-01-01'
end='2020-12-31'
#RECALL DATA FROM YAHOO FINANCE (APPLE, GOOGLE)
AAPL=yf.download("AAPL", start=start, end=end)
BABA= yf.download("BABA", start=start, end=end)
```

```python
#RECALL DATA FROM YAHOO FINANCE (GOOGLE)
GOOG= yf.download("GOOGL", start=start, end=end)
```

```python
#get first 5 rows and save data to csv files
AAPL.head()
AAPL.to_csv('apple_stock')
```
